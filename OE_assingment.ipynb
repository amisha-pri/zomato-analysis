**Zomato Restaurant Data** **Analysis**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("zomato.csv", encoding='latin-1')
```

```python
display(df.head(10))
```

```python
print(df.shape)
```

```python
print(df.columns)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
duplicate_rows = df.duplicated().sum()

print("Duplicate Rows:", duplicate_rows)
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.dtypes)
```

```python
df['Aggregate rating'] = pd.to_numeric(
    df['Aggregate rating']
)
```

```python
print(df.shape)
```

```python
top_cuisines = df['Cuisines'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_cuisines.values,
    y=top_cuisines.index,
    palette='magma'
)

plt.title(
    "Top 10 Cuisines",
    fontsize=15
)

plt.xlabel("Count")
plt.ylabel("Cuisine")

plt.grid(alpha=0.3)

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='Has Online delivery',
    data=df,
    palette='magma'
)

plt.title("Online Delivery Availability")

plt.show()
```

```python
sns.countplot(
    x='Has Table booking',
    data=df,
    palette='rocket'
)

plt.title("Table Booking Analysis")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Aggregate rating'],
    bins=20,
    kde=True,
    color='purple'
)

plt.title("Restaurant Rating Distribution")

plt.xlabel("Ratings")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df['Average Cost for two'],
    bins=30
)

plt.title("Average Cost for Two")

plt.xlabel("Cost")

plt.show()
```

```python
top_cuisines = df['Cuisines'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_cuisines.values,
    y=top_cuisines.index,
    palette='viridis'
)

plt.title("Top 10 Cuisines")

plt.xlabel("Count")
plt.ylabel("Cuisine")

plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
country = df['Country Code'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=country.values,
    y=country.index,
    palette='pastel'
)

plt.title(
    "Top Countries by Restaurants",
    fontsize=15
)

plt.xlabel("Count")
plt.ylabel("Country Code")

plt.grid(alpha=0.3)

plt.show()
```

```python
best_restaurants = df[
    ['Restaurant Name', 'Aggregate rating', 'Votes']
]

best_restaurants = best_restaurants.sort_values(
    by='Aggregate rating',
    ascending=False
)

display(best_restaurants.head(10))
```

```python
expensive = df[
    ['Restaurant Name', 'Average Cost for two']
]

expensive = expensive.sort_values(
    by='Average Cost for two',
    ascending=False
)

display(expensive.head(10))
```
